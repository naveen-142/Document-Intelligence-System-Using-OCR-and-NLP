```python
# Modules
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Pdf path
pdf_path = "/kaggle/input/ocrsamples/DryLab.pdf"
# Read
doc = fitz.open(pdf_path)
print(f"Opened PDF with {doc.page_count} pages")

all_text = ""

# Looping through Pages
for i, page in enumerate(doc):
    # Render page to image (300 DPI for good OCR) -> DPI: Dots Per Inch
    mat = fitz.Matrix(300/72, 300/72)  # scale to DPI 
    pix = page.get_pixmap(matrix=mat, alpha=False)

    # Convert Pixmap -> PIL Image
    img = Image.open(io.BytesIO(pix.tobytes("png")))

    # OCR with Tesseract
    text = pytesseract.image_to_string(img, lang="eng")
    all_text += f"\n--- Page {i+1} ---\n{text}"

doc.close()

print(all_text)
```

### 2. EasyOCR (Deep Learning-based)
ðŸ‘‰ Very easy to use, works well on natural scene text (signboards, street images).

* **On Images**

```python
# Module
import easyocr

# Initialize reader
reader = easyocr.Reader(['en'])  # supports multiple languages

# Run OCR
results = reader.readtext("/kaggle/input/ocrsamples/Vehicles.jpg")

# Print results
for (bbox, text, prob) in results:
    print(f"{text} (Confidence: {prob:.2f})")
```

* **On Documents**

```python
# Modules
import fitz  # PyMuPDF
import easyocr
from PIL import Image
import io
import numpy as np

# Reading pdf
pdf_path = "/kaggle/input/ocrsamples/DryLab.pdf"
doc = fitz.open(pdf_path)

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'], gpu=True)  # set gpu=True if you have GPU support

all_text = ""

for i, page in enumerate(doc):
    # Render page to image (300 DPI for good OCR)
    mat = fitz.Matrix(300/72, 300/72)
    pix = page.get_pixmap(matrix=mat, alpha=False)

    # Convert Pixmap -> PIL Image
    img = Image.open(io.BytesIO(pix.tobytes("png")))

    # Convert PIL -> NumPy array (RGB) for EasyOCR
    img_np = np.array(img.convert("RGB"))

    # OCR with EasyOCR
    results = reader.readtext(img_np)

    # Collect text
    page_text = " ".join([res[1] for res in results])
    all_text += f"\n--- Page {i+1} ---\n{page_text}"

doc.close()

print(all_text)
```

### 4. Google Cloud Vision OCR

ðŸ‘‰ Requires Google Cloud API key setup. Good for high accuracy & large-scale OCR.

```python
from google.cloud import vision

client = vision.ImageAnnotatorClient()

with open("sample.png", "rb") as img_file:
    content = img_file.read()

image = vision.Image(content=content)

response = client.text_detection(image=image)
for text in response.text_annotations:
    print(text.description)
```
